```python
#All together

import discord
from discord.ext import commands


#Partie Historique

class Node :
  def __init__(self, data):
    self.data = data
    self.next_node = None

class chained_list:
  def __init__(self):
    self.first_node = None
    self.last_node = None
    self.length = 0

  def append(self,data):
    new_node = Node(data)
    if self.first_node == None :
      self.first_node = Node(data)
      self.last_node = new_node
      self.length += 1
      return

    current_node = self.first_node
    while current_node.next_node != None:
      current_node = current_node.next_node

    current_node.next_node = new_node

    self.last_node = new_node
    self.length += 1

  def last_node(self):
    return self.last_node

  def length(self):
    return self.length

  def search(self, searched):
    if self.first_node == None :
      return False

    current_node = self.first_node
    while current_node.next_node != None:
      if current_node.data == searched :
        return True
      current_node = current_node.next_node
    return False

  def get(self, index):
    if self.first_node == None :
      return "Pas de valeur"

    i = 0
    current_node = self.first_node
    while current_node.next_node != None:
      if i == index :
        return current_node.data
      elif i + 1 == index :
        return current_node.next_node.data
      i += 1
      current_node = current_node.next_node
    return "Pas de valeur"

  # search qui dit oui ou non une valeur est présente
  # Get qui donne une valeur à un indice donné

  def insert(self, value, id):
    i = 0
    current_node = self.first_node
    while i != id : # current_node.next_node != None:
      current_node = current_node.next_node
      i += 1
    inserted_node = Node(value)
    inserted_node.next_node = current_node.next_node
    current_node.next_node = inserted_node

  def clearAll(self):
    self.first_node = None
    self.last_node = None
    self.length = 0

  def displayAllCommand(self):

    current_node = self.first_node
    display = current_node.data
    while current_node.next_node != None:
      current_node = current_node.next_node
      display += "\n" + current_node.data
    return display


#Instanciation Historique

Historique = chained_list()
index = 0


#Partie Arbre de Discussion

class Node2 :
  def __init__(self, message):
    self.message = message
    self.yes_node = None
    self.no_node = None

  def add_message(self, new_message, yes_or_no, old_message ):
    if self.message == old_message :
      if yes_or_no == "oui":
        self.yes_node = Node2(new_message)
      else:
        self.no_node = Node2(new_message)
    else:
      if self.yes_node != None:
        self.yes_node.add_message(new_message, yes_or_no, old_message )
      if self.no_node != None:
        self.no_node.add_message(new_message, yes_or_no, old_message )

class Discusion_tree:
  def __init__(self):
    self.first_node = None
    self.current_conversation_node = None
    self.parcour_node = None

  def next_message(self, yes_or_no):
    if yes_or_no == "yes":
      if self.current_conversation_node != None:
        self.current_conversation_node = self.current_conversation_node.yes_node
    else:
      if self.current_conversation_node != None:
        self.current_conversation_node = self.current_conversation_node.no_node


  def add_message(self, new_message, yes_or_no, old_message ):
    self.first_node.add_message(new_message, yes_or_no, old_message)

  def show_all_arbre(self, current_node):
    if current_node is not None:
      print(current_node.message)
      self.parcour_arbre(current_node.yes_node)
      self.parcour_arbre(current_node.no_node)

  def search_message(self, current_node, msgSearched):
    if current_node is not None:
      if msgSearched in current_node.message:
        return True
      found_in_yes = self.search_message(current_node.yes_node, msgSearched)
      found_in_no = self.search_message(current_node.no_node, msgSearched)
      return found_in_yes or found_in_no
    return False


#Instanciation Discution_tree

Discusion = Discusion_tree()
Discusion.first_node = Node2("Help ?")

Discusion.add_message("Python ?","oui","Help ?")
Discusion.add_message("Golang ?","non","Help ?")

Discusion.add_message("docs.python.org","oui","Python ?")
Discusion.add_message("Tant pis","non","Python ?")
Discusion.add_message("go.dev","oui","Golang ?")
Discusion.add_message("Humain ?","non","Golang ?")

Discusion.add_message("Sure ?", "oui", "Humain ?")
Discusion.add_message("Calling 911", "non", "Humain ?")

Discusion.add_message("Then I don't have other langages, sorry ¯\_(ツ)_/¯", "oui", "Sure ?")
Discusion.add_message("Calling 911", "non", "Sure ?")



#Partie stockage HashMap

def AppendIdHash(HashMap, keyId, value):
  index = 0
  for i in HashMap:
    if i[0] == keyId :
      HashMap[index] = (keyId, value)
      return True
    index += 1
  HashMap.append((keyId, value))
  return False

HashMap = []


#Partie Discord bot

intents = discord.Intents.all()

client = commands.Bot(command_prefix="!", intents = intents)

testbot_channel_token = 0000000000000000000 #METTRE ICI LE TOKEN D'UN CHANNEL POUR TESTER LE BOT

#Partie Commandes


#Historique

@client.command(name="Hello")
async def hello(ctx):
    Historique.append("Hello")

    testbot_channel = client.get_channel(testbot_channel_token)
    await testbot_channel.send("Hello")

@client.command(name="Suis_je_beau")
async def hello(ctx):
    Historique.append("Suis_je_beau")

    testbot_channel = client.get_channel(testbot_channel_token)
    await testbot_channel.send("Bien sûr")

@client.command(name="Dernierecommande")
async def lastCommand(ctx):
    testbot_channel = client.get_channel(testbot_channel_token)

    if Historique.last_node == None :
      await testbot_channel.send("Pas de derniere commande")
    elif Historique.last_node.data == "Dernierecommande":
      await testbot_channel.send("Ba tu viens d'appeler cette commande")
    else:
      await testbot_channel.send(Historique.last_node.data)

    Historique.append("Dernierecommande")

@client.command(name="Clearhistorique")
async def clearHisto(ctx):
    Historique.append("ClearHistorique")
    Historique.clearAll()

    testbot_channel = client.get_channel(testbot_channel_token)
    await testbot_channel.send("L'historique de commande a été vidé")

@client.command(name="Historiquecommande")
async def HistoriqueCommand(ctx):
    Historique.append("HistoriqueCommande")
    testbot_channel = client.get_channel(testbot_channel_token)

    if Historique.first_node != None :
      display = Historique.displayAllCommand()
      await testbot_channel.send(display)
    else:
      await testbot_channel.send("Aucune commande faite")

@client.command(name="HistoriqueMove")
async def HistoriqueMove(ctx):
    Historique.append("HistoriqueMove")
    testbot_channel = client.get_channel(testbot_channel_token)
    global index
    index = Historique.length - 1

    await testbot_channel.send("Commande à l'index " + str(index) + " : " + Historique.get(index))

@client.command(name="HistoriqueAvancer")
async def HistoriqueAvancer(ctx):
    Historique.append("HistoriqueAvancer")
    testbot_channel = client.get_channel(testbot_channel_token)

    global index
    if index != Historique.length:
      index += 1
      await testbot_channel.send("Commande à l'index " + str(index) + " : " + Historique.get(index))
    else:
      await testbot_channel.send("C'est déjà la dernière commande")

@client.command(name="HistoriqueReculer")
async def HistoriqueReculer(ctx):

    Historique.append("HistoriqueReculer")
    testbot_channel = client.get_channel(testbot_channel_token)

    global index
    if index != 0:
      index -= 1
      await testbot_channel.send("Commande à l'index " + str(index) + " : " + Historique.get(index))
    else:
      await testbot_channel.send("C'est déjà la première commande")


#Discussion_Tree + stockage HashMap id de l'utilisateur et avancement de la discussion

@client.command(name="Help")
async def help(ctx):

    Discusion.current_conversation_node = Discusion.first_node

    testbot_channel = client.get_channel(testbot_channel_token)
    await testbot_channel.send(Discusion.first_node.message)


    AppendIdHash(HashMap, ctx.author.global_name + ctx.author.name, Discusion.current_conversation_node.message)

    #TCommandes pour voir si la HashMap :
    #print(ctx.author.global_name + ctx.author.name)
    #print(HashMap)

@client.command(name="Yes")
async def yes(ctx):
    Discusion.next_message("yes")
    #print(Discusion.current_conversation_node.message)

    testbot_channel = client.get_channel(testbot_channel_token)
    if Discusion.current_conversation_node != None:
      await testbot_channel.send(Discusion.current_conversation_node.message)
      AppendIdHash(HashMap, ctx.author.global_name + ctx.author.name, Discusion.current_conversation_node.message)
    else:
      await testbot_channel.send("Fin de la discussion")

@client.command(name="No")
async def no(ctx):
    Discusion.next_message("no")
    #print(Discusion.current_conversation_node.message)

    testbot_channel = client.get_channel(testbot_channel_token)
    if Discusion.current_conversation_node != None:
      await testbot_channel.send(Discusion.current_conversation_node.message)
      AppendIdHash(HashMap, ctx.author.global_name + ctx.author.name, Discusion.current_conversation_node.message)
    else:
      await testbot_channel.send("Fin de la discussion")

@client.command(name="reset")
async def reset(ctx):
    Discusion.current_conversation_node = Discusion.first_node
    testbot_channel = client.get_channel(testbot_channel_token)
    await testbot_channel.send("La discussion a été reset")

    AppendIdHash(HashMap, ctx.author.global_name + ctx.author.name, Discusion.first_node.message)

@client.command(name="speakabout")
async def speak_about(ctx, *, subject=""):
    testbot_channel = client.get_channel(testbot_channel_token)

    if not subject:
      await testbot_channel.send("Vous n'avez pas mis de sujet")
      return

    if Discusion.search_message(Discusion.first_node, subject):
      await testbot_channel.send("Oui le bot parle de ce sujet")
    else:
      await testbot_channel.send("Non le bot ne parle pas de ce sujet")


#Partie exècution du bot

@client.event
async def on_ready():
    print("Le bot est prêt !")

    general_channel = client.get_channel(testbot_channel_token)
    await general_channel.send("------------------------------------------------------------------------\nJoel est opérationnel \nVoici les commandes pour l'historique : \n!Hello \n!Dernierecommande \n!Clearhistorique"
    "\n!Historiquecommande \nUtiliser d'abord HistoriqueMove pour parcourir l'historique" + "\n!HistoriqueMove \n!HistoriqueAvancer \n!HistoriqueReculer"
    + "\n\nVoici les commandes pour l'arbre de discussion (avec la HashMap) : \n!Help \n!Yes \n!No \n!reset \n!speakabout \n------------------------------------------------------------------------")

@client.event
async def on_member_join(member):
    general_channel = client.get_channel(testbot_channel_token)
    await general_channel.send("Bienvenue sur le serveur ! "+ member.name)

@client.event
async def on_message(message):
  if message.author == client.user:
    print("Il écrit le bot la") #Le bot envoie un msg
    return

  #message.content = message.content.lower()

  if message.content.startswith("Hello"):
    await message.channel.send("Hello")

  await client.process_commands(message)

client.run("TOKEN_DU_BOT")
```

```python
import discord
from discord.ext import commands
import requests

NASA_API_KEY = 'NASA_API_KEY'  # La clé API de la NASA qui est avec le lien du repo sur Moodle 
                                                                                                     # (NE PAS PARTAGER !)

#client = discord.Client()
intents = discord.Intents.all()

client = commands.Bot(command_prefix="!", intents = intents)

testbot_channel_token = 0000000000000000000 #METTRE ICI LE TOKEN D'UN CHANNEL POUR TESTER LE BOT
                        
@client.command(name="nasaPlanetOfTheDay")
async def nasaPlanetOfTheDay(ctx):
    testbot_channel = client.get_channel(testbot_channel_token)

    try:
      response = requests.get(f'https://api.nasa.gov/planetary/apod?api_key={NASA_API_KEY}')
      data = response.json()

      embed = discord.Embed(title=data['title'], description=data['explanation'], color=0x00ff00)
      embed.set_image(url=data['url'])

      await testbot_channel.send(embed=embed)

    except Exception as e:
       await testbot_channel.send(f"Une erreur s'est produite : {e}")

@client.command(name="nasaAsteroids")
async def nasaAsteroids(ctx):
    testbot_channel = client.get_channel(testbot_channel_token)
    try:
      response = requests.get(f'https://api.nasa.gov/neo/rest/v1/feed/today?detailed=false&api_key={NASA_API_KEY}')
      data = response.json()

      near_earth_objects = data['near_earth_objects']
      embed = discord.Embed(title="Astéroïdes proches de la Terre", color=0xff0000)

      for date in near_earth_objects:
        for asteroid in near_earth_objects[date]:
          asteroid_name = asteroid['name']
          asteroid_diameter = asteroid['estimated_diameter']['kilometers']['estimated_diameter_max']
          asteroid_close_approach = asteroid['close_approach_data'][0]['close_approach_date']
                    
          embed.add_field(name=asteroid_name, value=f"Diamètre estimé: {asteroid_diameter} km\nDate de l'approche: {asteroid_close_approach}", inline=True)

      await testbot_channel.send(embed=embed)

    except Exception as e:
        await testbot_channel.send(f"Une erreur s'est produite : {e}")
        print(f"Une erreur s'est produite : {e}")

@client.command(name="nasaMarsPhotos")
async def nasaMarsPhotos(ctx, *, subject=""):
    testbot_channel = client.get_channel(testbot_channel_token)
    try:
      if subject == "" :
        IndexInPhotosDB = 0
        await testbot_channel.send("Vous pouvez aussi spécifié la photo voulu en ajoutant un nombre après la commande (de 0 à 855), en attendant voici la 1ère photo du rover curiosity :")
      elif subject.isdigit():
        IndexInPhotosDB = int(subject)
      else:
        await testbot_channel.send("La commande n'accepte que des nombres positifs et entier entre 0 et 855 pour choisir la photo")
        return
            
      response = requests.get(f'https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?sol=1000&api_key={NASA_API_KEY}')
      data = response.json()

      photos = data['photos']
      embed = discord.Embed(title="Photos récentes prises par Curiosity sur Mars", color=0xff0000) 
      
      photos = data['photos']
            
      if photos:
        photo = photos[IndexInPhotosDB]
        img_src = photo['img_src']
        earth_date = photo['earth_date']
        camera = photo['camera']['full_name']

        embed = discord.Embed(title="Photo récente prise par Curiosity sur Mars", color=0xff0000)
        embed.add_field(name=f"Date terrestre: {earth_date}", value=f"Caméra: {camera}", inline=False)
        embed.set_image(url=img_src)

      await testbot_channel.send(embed=embed)

    except Exception as e:
      await testbot_channel.send(f"Une erreur s'est produite : {e}")

@client.command(name="nasaISS") 
async def nasaISS(ctx):
    testbot_channel = client.get_channel(testbot_channel_token)
    try:
      response_iss = requests.get('http://api.open-notify.org/iss-now.json')
      data_iss = response_iss.json()

      timestamp = data_iss['timestamp']
      position = data_iss['iss_position']

      embed = discord.Embed(title="Position actuelle de l'ISS", color=0xff0000)
      embed.add_field(name="Timestamp", value=timestamp, inline=False)
      embed.add_field(name="Latitude", value=position['latitude'], inline=True)
      embed.add_field(name="Longitude", value=position['longitude'], inline=True)

      await testbot_channel.send(embed=embed)

    except Exception as e:
      await testbot_channel.send(f"Une erreur s'est produite : {e}")

@client.command(name="nasaAstronauts") 
async def nasaAstronauts(ctx):
    testbot_channel = client.get_channel(testbot_channel_token)
    try:
      response_astronauts = requests.get('http://api.open-notify.org/astros.json')
      data_astronauts = response_astronauts.json()

      embed = discord.Embed(title="Astronautes actuellement dans l'espace", color=0xff0000)      
      embed.add_field(name="Astronautes en orbite", value=f"{data_astronauts['number']} astronautes", inline=False)

      for astronaut in data_astronauts['people']:
        embed.add_field(name=astronaut['name'], value=astronaut['craft'], inline=True)

      await testbot_channel.send(embed=embed)

    except Exception as e:
      await testbot_channel.send(f"Une erreur s'est produite : {e}")

@client.command(name="nasaMarsWeather")
async def nasaMarsWeather(ctx, *, subject=""):
    testbot_channel = client.get_channel(testbot_channel_token)
    try:
      response = requests.get('https://api.maas2.apollorion.com/')
      data = response.json()

      date = data['terrestrial_date']
      martianSol = data['sol']
      min_temp = data['min_temp']
      max_temp = data['max_temp']
      pressure = data['pressure']
      sunrise = data['sunrise']
      sunset = data['sunset']

      embed = discord.Embed(title="Données météorologiques récentes sur Mars", color=0xff0000)

      embed.add_field(name="Date de référence Terrestre", value=date, inline=False)
      embed.add_field(name="Sol martien (date)", value=martianSol, inline=False)
      embed.add_field(name="Température minimum", value=min_temp, inline=True)
      embed.add_field(name="Température maximum", value=max_temp, inline=True)
      embed.add_field(name="Pression", value=pressure, inline=True)
      embed.add_field(name="Heure du lever de Soleil", value=sunrise, inline=False)
      embed.add_field(name="Heure du coucher de Soleil", value=sunset, inline=False)

      await testbot_channel.send(embed=embed)

    except Exception as e:
      await testbot_channel.send(f"Une erreur s'est produite : {e}")

@client.event
async def on_ready():
    print("Le bot est prêt")

    general_channel = client.get_channel(testbot_channel_token)
    await general_channel.send("------------------------------------------------------------------------\nJoel est opérationnel \nVoici les commandes pour les informations sur l'espace :"
    + "\nnasaPlanetOfTheDay : Photo/description journalière d'un object céleste"
    + "\nnasaAsteroids : Tout les asteroids actuellement proche de la Terre"
    + "\nnasaMarsPhotos : Photos de Mars du rover Curiosity (de 0 à 855)"
    + "\nnasaISS : Position actuel de l'ISS dans l'espace"
    + "\nnasaAstronauts : Astronautes actuellement dans l'espace"
    + "\nnasaMarsWeather : Description de la météo de Mars")

@client.event
async def on_message(message):
    if message.author == client.user:
        return

    if message.author == client.user:
      print("Il écrit le bot la") #Le bot envoie un msg
      return

    #message.content = message.content.lower()

    if message.content.startswith("Hello"):
      await message.channel.send("Hello")

    await client.process_commands(message)


client.run("TOKEN_DU_BOT")
```
